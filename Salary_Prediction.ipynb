```python
mean_value = df['Experience Years'].mean()
mean_value_sal=df['Salary'].mean()
df['Experience Years'] = df['Experience Years'].fillna(value=mean_value)
df['Salary']=df['Salary'].fillna(value=mean_value_sal)
```

```python
#split the data in dependent and independent variable
X=df.iloc[:,:-1]
y=df[["Experience Years"]]
```

```python
X
```

```python
y
```

```python
#training: model building 
#test: prediction
##Performin train test split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.75,random_state=42, shuffle=True)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
#Building linear regression model
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
#y=mx+c
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred_tr=lr.predict(X_train)
y_pred_te=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_pred_te)
print("mean absolute error : ",mae)
```

```python
from sklearn.metrics import r2_score
```

```python
print("Train accuracy",r2_score(y_train,y_pred_tr))
print("**"*25)
print("Test accuracy",r2_score(y_test,y_pred_te))
```

```python
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, y_pred_tr, color='red', label='Regression line')
```

```python
plt.scatter(X_test, y_test, color='green', label='Testing data')
plt.plot(X_test, y_pred_te, color='red', label='Regression line')
```
